## image registration
semi-manual: 
1. overlay in powerpoint with alpha=0.5 
2. rotate/rescale until they look to overlay to the eye
3. apply rotation/rescale in this notebook and adjust overlay params
4. check alignment in zoomed in frames

```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
```

```python
he_slide_path="/Users/rahma/bf_bottomright.jpg"
dt_slide_path="/Users/rahma/polyt_bottomright_invert.jpg"

Image.MAX_IMAGE_PIXEL = None
he = np.array(Image.open(he_slide_path))
dt = np.array(Image.open(dt_slide_path))
```

```python
plt.figure(dpi=300)

plt.imshow(he, cmap='gray', alpha=0.5)
plt.imshow(dt, cmap='gray',  alpha=0.5)

plt.show();
```

Try rescaling according to manual rescaling done in powerpoint. Only the polyT had rotation/rescale adjustment

```python
dt = Image.open(dt_slide_path)
dt_width, dt_height = dt.size
dt_width, dt_height
```

```python
# rescale factor
dt_hs = int(7.96/7.5 * dt_height)
dt_ws = int(6.77/6.35 * dt_width)
```

```python
#rescale
newsize = (dt_ws, dt_hs)
dt_rescaled = dt.resize(newsize)

plt.imshow(np.array(dt_rescaled), cmap='gray')
```

```python
# rotate
dt_rot = dt_rescaled.rotate(1)
```

```python
plt.figure(dpi=200)

plt.imshow(np.array(he), cmap='gray', alpha=0.5)
plt.imshow(np.array(dt_rot), cmap='gray', alpha=0.5)

plt.show();
```
